### Task 2.4: Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
import matplotlib.pyplot as plt

# Drop non-date columns before grouping
df_clean = df.drop(columns=["Lat", "Long", "Province/State"], errors='ignore')

# Group by Country/Region and sum all date columns
corona_dataset_aggregated = df_clean.groupby("Country/Region").sum()

# Transpose so that dates are the index and countries are columns
transposed_df = corona_dataset_aggregated.T

# Optional: check available country names
print(transposed_df.columns.tolist()[:10])  # Show first 10 country names

# Now plot the desired countries
transposed_df["China"].plot()
transposed_df["Italy"].plot()
transposed_df["Spain"].plot()
transposed_df["Turkey"].plot()

plt.legend(["China", "Italy", "Spain", "Turkey"])
plt.title("Covid-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### task 3.2: find maxmimum infection rate for China

```python
corona_dataset_aggreageted.loc["Turkey"].diff().max()
```

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
```

```python
corona_dataset_aggreageted["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggreageted.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggreageted["max_infection_rate"])
```
